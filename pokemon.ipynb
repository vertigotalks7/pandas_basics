# Pandas tutorial


```python
import pandas as pd
df = pd.read_csv("sample_data/pokemon_data.csv")
#df = pd.read_excel("sample_data/pokemon_data.xlsx")
#df = pd.read_csv("sample_data/pokemon_data.txt",delimiter='\t')


print(df.head(3))


#print(df.tail(3))
```

reading data in pandas

```python
#read all columns
print(df.columns)

#read a specific column/columns
print(df[['Name','Type 1','HP']][0:10])
print("\n")

#print each row
print(df.iloc[0:4])
print("\n")

#print specific location
print(df.iloc[4,1]) #df.iloc[row,column]


```

**iteration o/p**

```python
for index,row in df.iterrows():
  print(index,row["Name"])
```

specific cell with matching data like "WHERE" condn in sql

check a cell which is not an integer with loc (i.e str,nums etc)

```python
df.loc[df["Type 1"]=="Grass"]
```

# Sorting/Describing data


```python
df.sort_values(['Type 1','HP'],ascending=[1,0])
```

#Making changes to the data

```python
#df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
#df = df.drop(columns=['Total'])
df['Total'] = df.iloc[:,4:10].sum(axis=1)

cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]

#df = df[['Total','HP','Defense']]
df.head(5)
```

saving it to csv

```python
df.to_csv("modified.csv",index=False)
#df.to_excel("modified.xlsx",index=False)
df.to_csv("modified.txt",index=False,sep='\t')
```
